### Usando o describe()

```python
df.describe() # count, mean, std, min, etc

df.sort_values('Name', ascending=False) # ordena de forma descendente o Nome

df.sort_values(['Type 1', 'HP'], ascending=True) # ordena de forma ascendente o type e depois o HP

df.sort_values(['Type 1', 'HP'], ascending=[1,0])[0:2] # o primeiro vai ser ascendente e o segundo termo será descendente
#ascending = vc consegue ver o menor HP
#descending = vc consegue ver o maior HP
```

### Mudando os dados

```python
#SOMA CRIANDO NOVA COLUNA
df['Total'] = df['HP'] +df['Attack'] +df['Defense'] +df['Sp. Atk'] +df['Sp. Def'] +df['Speed'] # vai criar uma nova coluna no dataframe chamada total e essa coluna vai ser o resultado dos parametros que vc colocou 

#EXCLUINDO COLUNA
df = df.drop(columns=['Total']) # exclui a coluna

#SOMA COM NOVA COLUNA E COM ILOC
#axis=1 soma horizontal, axis=0 soma vertical
df['Total'] = df.iloc[:,4:10].sum(axis=1) # primeiro termo ':' significa todas as linhas, '4:9' vai partir da 4 coluna até a 9 coluna,sum vai somar elas 

df.head(4)
```

### Filtrando os Dados

```python
df.loc[df['Type 1'] == 'Grass'] #vai mostrar apenas as linhas de type que tiverem o valor 'Glass'

df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')] # & = and

df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]

df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')] # | = or
```

### Novo dataframe com um filtro criado

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)] # & = and


# baixar o novo dataframe na forma de csv
#new_df.to_csv('filtered.csv', index=False) 

new_df.reset_index(drop=True, inplace=True)  #drop = True vai fazer com que ele drope o index antigo. #inplace=True vai coloca um novo indice no seu dataframe
new_df
```

### Regex Filtering

```python
#filtrando 'nomes' que contenham alguma 'palavra' de interesse

df.loc[df['Name'].str.contains('Mega')] #vai encontrar todos os nomes que contenham Mega

df.loc[~df['Name'].str.contains('Mega')][0:5] #vai encontrar todos os nomes que NÃO contenham Mega

```

### RegEx complexos

```python
import re # importante expressoes regulares = RegEx (Regular Expression)

df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)] # encontra o type 1 que tiverem: fire ou glass

df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)] #flags=re.I vai ignorar a letra minuscula e vai mostrar os resultados --> realizar a correspondência que não diferencia maiúsculas de minúsculas. Isso significa que o [A-Z]também corresponderá a letras minúsculas.

df.loc[df['Name'].str.contains('pi[a-z]*', flags=re.I, regex=True)][0:5] #encontra tudo que tiver pi[qualquer letra de 'a' até 'z']; o '*' significa um ou mais
```

### Condicional

```python
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer' #vai mudar fire para flamer. 'Type 1' no final vai especificar a coluna que sera mudada

df.loc[df['Type 1'] == 'Fire', 'Legendary'] = True # vai colocar o valor true na coluna legendary onde Type 1 = Fire


df.loc[df['Total'] > 500, ['Generation', 'Legendary']] = 'TEST VALUE' #onde tiver a coluna total maior que 500, vai mudar a coluna legendary e generation para o valor que vc atribuir ( != diferente)


df.loc[df['Total'] > 500, ['Generation', 'Legendary']] = ['teste 1', 'test 2'] #coloca test1 na coluna do primeiro termo e test2 na coluna do segundo termo

df[0:5]
```

### Funções de agregação estatística (Groupby, mean, sum, count)


```python
df.groupby(['Type 1']).mean() # vai agrupar pelo Type 1 e depois vai tirar a média de cada tipo (levando em consideracao apenas os valores numericos)

df.groupby(['Type 1']).mean().sort_values('Defense') #vai fazer a media e ordenar pela coluna Defense

df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False) #vai fazer a media de forma descendente

df.groupby(['Type 1']).sum() #vai somar os valores agrupando pelo 'type 1'

df.groupby(['Type 1']).count() # conta a quantidade de valores de cada coluna (se tiver em branco ele nao conta)

#criando uma nova coluna para fazer um count
df['count'] = 1 #cria uma coluna onde todas as linhas serão preenchidas com '1'
df.groupby(['Type 1']).count()['count'] #vai contar agrupando a coluna type1 e o '['count']' no final faz com que só apareça a coluna count

#agrupando varias colunas
df.groupby(['Type 1', 'Type 2']).count()['count']
```
