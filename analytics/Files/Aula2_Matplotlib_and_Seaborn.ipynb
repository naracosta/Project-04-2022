### Loop graph

```python
#for country in gas:
#    print(country) #vai printar todas as colunas do dataframe gas

plt.figure(figsize=(10,5)) #aumenta o tamanho do graph (eixo x: 10 ; eixo y: 5)

plt.title('Gas Prices over Time (in USD)') # titulo do graph

for country in gas:
    if country != 'Year':
        #print(country) #vai printar todas os paises do dataframe gas, sem contar a coluna de Year
        plt.plot(gas.Year, gas[country], marker='.', label = country) #vai plotar um grafico com o seu loop e colocar pontos nas linhas, com label igual a cada país

        
        plt.xticks(gas.Year[::3])
plt.legend()
```

### Histogram 


```python
fifa = pd.read_csv(r"C:\Users\aluga.com\Documents\meus_projetos\Project-DataScience\analytics\Files\dataframes\matplotlib_tutorial-master\fifa_data.csv")
fifa.head(3)
```

```python
bins = [40,50,60,70,80,90,100] #intervalo que vc quer no eixo x

plt.hist(fifa.Overall, bins=bins, color='#d68c87') #overall é a coluna de interesse, bins é o intervalo do eixo x, color é a cor do histograma

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribuition of Player Skills in FIFA 2018')



plt.show()
```

### Pie Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0] #faz o count levando em conta a coluna e o valor letf que esta na coluna
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left','Right']
colors = ['#d68c87','#aabbcc']
plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Preferred Foot of FIFA Players')

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight] # vai remover o lvs no final de cada valor da coluna weight

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy =  fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125','125-150','150,175','175-200','Over 200']
explode = (.4,.1,0,0,.4) # vai separa o pie graph

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance = 0.8, explode = explode )

plt.title('Weight Distribution of FIFA Players (in lbs)')

plt.show()
```

```python
sns.displot(df['not_distracted'], kde=False, bins=25) # y:count; x: 'not_distracted', bins é o tamanho do graph ; kde=normaliza
```

### Join Plot

```python
plt.figure(figsize=(19,4))
sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg') #vai adicionar no seu grafico - data= 'seu dataframe', e x e y sao as colunas do seu dataframe
```

```python
#dataframe que mostra a distribuição de gorjeta
tips_df = sns.load_dataset('tips') #abre o datafram tips
print(tips_df.head(2)) #conferindo as linhas do dataframe
sns.pairplot(tips_df, hue='sex', palette='Blues') #plot distribuição da coluna Sex para os outros parametros (com a cor azul)
```

### Rug Plot

```python
sns.rugplot(tips_df['tip']) 
```

### Bar Plot

```python
sns.barplot(x='sex', y='total_bill', data=tips_df) #data=datafram_de_que_vc_esta_usando. Os dados serão agrupados pela média

sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.median) #dados agrupados pela mediana

sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.std) #dados agrupados pela desvio padrão

sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.var) #dados agrupados pela variancia

sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.cov) #dados agrupados pela covariancia
```

### Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex') #distribuição da quantidade de gorjeta que uma mulher ganha por dia
#hue agrupa por sexo
```
